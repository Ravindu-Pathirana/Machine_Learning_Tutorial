```python
A= (5,3)
B=(2,4)
```

```python
from scipy.spatial import distance
```

### Euclidean Distance

```python
euclidean_distance=distance.euclidean(A,B)
print(euclidean_distance)
```

### Manhatten Distance

```python
manhatten_distance=distance.cityblock(A,B)
print(manhatten_distance)
```

### Cosine Distance

```python
cosine_diatnace=distance.cosine(A,B)
print(cosine_diatnace)
```
