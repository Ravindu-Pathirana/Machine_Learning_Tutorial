```python
import pandas as pd
data = pd.read_csv('Student.csv')
print(data)
```

```python
result_category=data['result']
```

```python
from sklearn.preprocessing import LabelEncoder
object=LabelEncoder()
result=object.fit_transform(result_category)
print(result)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
result1=ohe.fit_transform(result_category.values.reshape(-1,1))
result1.toarray()
```

```python
from sklearn.preprocessing import LabelBinarizer
lb=LabelBinarizer()
result2=lb.fit_transform(result_category)
print(result2)
```
