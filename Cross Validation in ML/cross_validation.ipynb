```python
import pandas as pd
data=pd.read_csv('Iris.csv')
data=data.drop("Id",axis=1)
```

```python
x=data.drop("Species",axis=1)
x.head()
```

```python
y=data["Species"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.model_selection import cross_val_score
```

### Cross Validation in Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100)
scores=cross_val_score(model,x_train,y_train,cv=5)
print("Cross Validation Scores:",scores)
print("Average Cross Validation Score:",scores.mean())
```

### Cross validation in K-NN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_scores=cross_val_score(knn_model,x_train,y_train,cv=5)
print("KNN Cross Validation Scores:",knn_scores)
print("KNN Average Cross Validation Score:",knn_scores.mean())
```

### Cross Validation in naive bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb_model=GaussianNB()
gnb_scores=cross_val_score(gnb_model,x_train,y_train,cv=5)
print("GaussianNB Cross Validation Scores:",gnb_scores)
print("GaussianNB Average Cross Validation Score:",gnb_scores.mean())
```

### Cross validation in SVM

```python
from sklearn.svm import SVC
svm_model=SVC(kernel='linear')
svm_scores=cross_val_score(svm_model,x_train,y_train,cv=5)
print("SVM Cross Validation Scores:",svm_scores)
print("SVM Average Cross Validation Score:",svm_scores.mean())
```
