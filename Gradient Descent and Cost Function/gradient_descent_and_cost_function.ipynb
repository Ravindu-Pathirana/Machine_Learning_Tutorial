```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
y= np.array([3,6,4,8,9,12,16])
x= np.array([5,6,7,8,9,10,11])
```

```python
m=0
b=0
alpha=0.01 # learning rate
iterations=1000
n=len(x)
```

```python
y_pred=m*x + b
y_pred
```

```python
for i in range(iterations):
    y_pred=m*x + b
    cost=sum((y - y_pred) ** 2) / n  # Mean Squared Error

    plt.scatter(m,cost)
    plt.xlabel('m')
    plt.ylabel('Cost')
    plt.title('Cost vs m during Gradient Descent')
    

    D_m = (-2/n) * sum(x * (y - y_pred))  # Derivative wrt m
    D_b = (-2/n) * sum(y - y_pred)        # Derivative wrt b

    #update parameters
    m = m - alpha * D_m
    b = b - alpha * D_b
    print(f"Iteration {i+1}: m={m}, b={b}, cost={cost}")

print(f"Final parameters: m={m}, b={b}")
plt.show()
```
