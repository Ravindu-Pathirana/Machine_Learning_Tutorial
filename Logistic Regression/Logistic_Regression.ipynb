```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load dataset
data = pd.read_csv('sample.csv')
```

```python
data
```

```python
plt.scatter(data['age'],data['job'])
plt.xlabel('Age')
plt.ylabel('Job')
plt.title('Age vs Job')
plt.show()
```

```python
m,c = np.polyfit(data['age'],data['job'],1)
plt.xlabel('Age')
plt.ylabel('Job')
plt.title('Age vs Job with Regression Line')
plt.scatter(data['age'],data['job'])
plt.plot(data['age'],m*data['age']+c,color='red')
plt.show()
```

We use the sigmoid function to map predicted values to probabilities in logistic regression.


```python
x = data[['age']]
x 
```

```python
y = data['job']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train
len(x_train)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
x_test
```

```python
y_test
```

```python
model.score(x_test, y_test)
```
