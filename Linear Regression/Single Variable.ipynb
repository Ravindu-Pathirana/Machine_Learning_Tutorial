```python
%pip install numpy pandas matplotlib scikit-learn
import numpy as np
import pandas as pd

## to visualization purposes
import matplotlib.pyplot as plt

## to build liner model
from sklearn.linear_model import LinearRegression

```

```python
## Load Dataset
data =pd.read_csv('Book.csv')
data
```

```python
## view data points
plt.scatter(data.videos,data.views,color='blue',marker='o',s=30)
plt.xlabel('Number of Videos')
plt.ylabel('Number of Views')
plt.title('Videos vs Views')
plt.show()
```

```python
data.videos
```

```python
data.views
```

```python
## divide data set into x and y
x=np.array(data.videos.values)
y=np.array(data.views.values)
```

```python
## build and train model
model=LinearRegression()
model.fit(x.reshape(-1,1),y)
```

```python
## Assingn a value to predict
num_videos=25
new_x=np.array(num_videos).reshape(-1,1)
predicted_views=model.predict(new_x)
print(f"Predicted views for {num_videos} videos: {predicted_views[0]}")
```

```python
## Assingn a value to predict
num_videos=45
new_x=np.array(num_videos).reshape(-1,1)
predicted_views=model.predict(new_x)
print(f"Predicted views for {num_videos} videos: {predicted_views[0]}")
```

```python
plt.scatter(data.videos,data.views,color='blue',marker='o',s=30)
m,c = np.polyfit(x, y, 1)
plt.plot(x, m*x + c, color='red')
plt.xlabel('Number of Videos')
plt.ylabel('Number of Views')
plt.title('Linear Regression Fit')
```
