```python
%pip install numpy pandas matplotlib scikit-learn
import numpy as np
import pandas as pd

## to visualization purposes
import matplotlib.pyplot as plt

## to build liner model
from sklearn.linear_model import LinearRegression
```

```python
# Load dataset
data = pd.read_csv('Book2.csv')
data
```

```python
# Build and Train Model

model=LinearRegression()
model.fit(data[['videos','days','subscribers']],data[['views']])
```

```python
## get prediction for new data point
model.predict([[45,180,3100]])
```

```python
plt.plot(data.videos,model.predict(data[['videos','days','subscribers']]),color='red')
plt.xlabel('Number of Videos')
plt.ylabel('Number of Views')
plt.show()
```
