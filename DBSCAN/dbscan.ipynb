```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# to generate datasets
from sklearn.datasets import make_blobs
from sklearn.datasets import make_circles

# to feature scaling
from sklearn.preprocessing import StandardScaler

# to build different clustering algorithms
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

```python
def display(X, n_clusters):
    plt.figure(figsize=(20,4))

    plt.subplot(1, 4, 1)
    plt.scatter(X[:,0], X[:,1])
    plt.xticks([])
    plt.yticks([])
    plt.title("Original Dataset")

    plt.subplot(1, 4, 2)
    y_pred = KMeans(n_clusters = n_clusters,random_state=0,n_init=200).fit_predict(X)
    plt.scatter(X[:,0], X[:,1], c=y_pred)
    plt.xticks([])
    plt.yticks([])
    plt.title("K-Means")

    plt.subplot(1, 4, 3)
    y_pred = AgglomerativeClustering(n_clusters=n_clusters).fit_predict(X)
    plt.scatter(X[:,0], X[:,1], c=y_pred)
    plt.xticks([])
    plt.yticks([])
    plt.title("Hierarchical")

    plt.subplot(1, 4, 4)
    y_pred = DBSCAN(eps=0.3, min_samples=10).fit_predict(X)
    plt.scatter(X[:,0], X[:,1], c=y_pred)
    plt.xticks([])
    plt.yticks([])
    plt.title("DBSCAN")
    plt.show()
```

```python
centers = [[1, 1], [-1, -1], [1, -1]]
X, _ = make_blobs(n_samples=750, cluster_std=0.4, centers=centers, random_state=0)
X = StandardScaler().fit_transform(X)
display(X, 3)
```

```python
X, _ = make_circles(n_samples=750, factor=0.3, noise=0.1, random_state=0)
X = StandardScaler().fit_transform(X)
display(X, 2)
```
